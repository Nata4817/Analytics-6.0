1. Використовуючи дані компаніі Yulu дослідіть, яка різниця між середніми показниками вологості по сезонам - побудуємо стовпачасті діаграми аби це дослідити.

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('Data/yulu_rental.csv')
seasonal_humidity = df.groupby('season')['humidity'].mean().reset_index()

# Перетворюємо числові значення сезонів на назви
seasonal_humidity['season'] = seasonal_humidity['season'].map({
    1: 'Весна', 
    2: 'Літо', 
    3: 'Осінь', 
    4: 'Зима'
})

# Створюємо стовпчасту діаграму
plt.figure(figsize=(8, 6))
plt.bar(seasonal_humidity['season'], seasonal_humidity['humidity'], color=['#4CAF50','#FF5722', '#FFC107', '#2196F3'])

plt.title('Середні показники вологості по сезонах', fontsize=16)
plt.xlabel('Сезон', fontsize=12)
plt.ylabel('Середня вологість (%)', fontsize=12)
plt.show()
```

Висновок:
Цей графік дозволяє легко побачити, в яких сезонах спостерігаються вищі чи нижчі показники вологості. Це може бути корисним для розуміння, як погода впливає на попит на прокат велосипедів у різні сезони.

2. Використовуючи файл yulu_rental.csv напиши код та побудувати графік залежності кількості взятих в аренду велосипедів від швидкості вітру і пори року

```python
if 'count' in df.columns and 'windspeed' in df.columns and 'season' in df.columns:
    # Визначення кольорів для кожного сезону
    season_colors = {
        1: 'spring',  # весна
        2: 'summer',  # літо
        3: 'fall',    # осінь
        4: 'winter'   # зима
    }
plt.figure(figsize=(12, 8))
    
    # Проходження через кожен сезон і побудова графіка
for season, color in season_colors.items():
    subset = df[df['season'] == season]
    plt.scatter(subset['windspeed'], subset['count'], label=f'Season {season}', alpha=0.2)

plt.title('Залежність кількості взятих в аренду велосипедів від швидкості вітру і сезону')
plt.xlabel('Швидкість вітру (м/с)')
plt.ylabel('Кількість взятих в аренду велосипедів')
plt.legend(title='Сезон')
plt.grid(True)
plt.show()
```

3. Побудуйте графік типу "ящик з вусами" для кількості прокатів по сезонам

```python
import seaborn as sns
# Створення графіку box plot з кольоровими сезонами
plt.figure(figsize=(10, 6))
sns.boxplot(x='season', y='count', data=df, hue='season', palette='Set2')

plt.title('Розподіл кількості прокатів по сезонам')
plt.xlabel('Сезон')
plt.ylabel('Кількість прокатів')
plt.legend(title='Сезони', loc='upper right')
plt.show()

```

Висновок:
За допомогою цього графіка можемо швидко побачити, як кількість прокатів змінюється в залежності від сезону. Якщо один сезон, наприклад, літо, показує більшу кількість прокатів, це може бути корисною інформацією для прийняття рішень щодо бізнесу або управління ресурсами.
